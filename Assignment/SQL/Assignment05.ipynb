```python
import mysql.connector as connection
```

```python
def executeQuery(query):
    con = connection.connect(host='localhost',user='root',passwd='admin',database='sakila',use_pure=True)
    cursor = con.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    con.close()
    return result
```

Q-1. Write an SQL query to print details of workers excluding first names, “Vipul” and
“Satish” from Worker table.

```python
query = "select * from worker where first_name not in('Satish','Vipul')"
executeQuery(query)
```

Q-2. Write an SQL query to print details of the Workers whose FIRST_NAME ends with

```python
query = "SELECT  first_name, last_name FROM worker WHERE length(first_name) = 5 and first_name like '%h';"
print("6 char and end with n ",executeQuery(query))
print("\n\n\n")
query = "SELECT  first_name, last_name FROM worker WHERE length(first_name) = 5 and first_name like '%k';"
print("6 char and end with k ",executeQuery(query))
```
